```python
def initialisation(file):
  Clients={}
  SatisfiedClients={}
  Propects={}
  file = open(file,'r')
  text=file.read()

  file.close()
  
  text=text.split('\n')
  N=text.pop(0)
  j=0
  for i in range(0,len(text)-1,2): 
      j+=1
      like=text[i][2:].split(' ')
      dislike=text[i+1][2:].split(' ')

      if dislike == ['']:
        dislike=[]

      Clients[j]=[like,dislike]

      if dislike == []:
        SatisfiedClients[j] = Clients[j]
      
      else:
          Propects[j] = Clients[j]
  
  return Clients, SatisfiedClients , Propects

```

```python
def extractIngredientList(Clients):
  Ingredients=[]
  for key,val in Clients.items():
    Ingredients+=val[0]

  return list(set(Ingredients))
```

```python
def dispatchClients(LikedIngredients,DislikedIngredients,UnsatisfyClients,SatisfiedClients,Prospects,n):

  for key,val in Prospects.items():
    if len(val[1]) == n :
      if set(val[1]).issubset(LikedIngredients):
        UnsatisfyClients[key]=val
      else:
        
        if list(set(val[0]) &  set(DislikedIngredients)) != []:
          UnsatisfyClients[key]=val
          
        else:
          SatisfiedClients[key]=val
          LikedIngredients+=val[0]
          DislikedIngredients+=val[1]

 
```

```python
def dislikeIngredientNumber(Prospects):
  L=[]
  for key,val in Prospects.items():
    L.append(len(val[1]))
    
  return list(set(L))


```

```python
#Programe principale

inputfile='/content/d_difficult.in.txt'

Clients,SatisfiedClients,Prospects = initialisation(inputfile)

UnsatisfyClients={}

DislikedIngredients=[]
LikedIngredients = extractIngredientList(SatisfiedClients)
# print(" ")
# print(len(LikedIngredients)," ", LikedIngredients)

# print(" UnsatisfyClients", UnsatisfyClients)
# print(" Clients", Clients)
# print(" Prospects", Prospects)
# print(" SatisfiedClients", SatisfiedClients)

for key,val in Prospects.items():
 
  if set(val[1]).issubset(LikedIngredients):
    UnsatisfyClients[key]=val
  else:
    
    if list(set(val[0]) &  set(DislikedIngredients)) != []:
      UnsatisfyClients[key]=val
      
    else:
      SatisfiedClients[key]=val
      LikedIngredients+=val[0]
      DislikedIngredients+=val[1]
      
    

LikedIngredients = extractIngredientList(SatisfiedClients)

print(len(LikedIngredients)," ", ' '.join(LikedIngredients))

print(" UnsatisfyClients ",len(UnsatisfyClients),"\n", UnsatisfyClients)

print(" SatisfiedClients",len(SatisfiedClients),"\n", SatisfiedClients)

print(" Clients", Clients)
print(" Prospects", Prospects)
print(" DislikeIngredient", DislikedIngredients)

L=[i for i in (set(SatisfiedClients.keys()) & set(UnsatisfyClients.keys()))]
print("Il y a ",len(L)," Clients repet√©: \n ",L)
        
print("Dislike Numbers: ",dislikeIngredientNumber(Prospects))
    

        

```
